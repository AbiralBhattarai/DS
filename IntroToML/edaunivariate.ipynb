```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/titanic.csv')
```

```python
df.sample(5)
```

Categorical Data

a. Count plot

```python
sns.countplot(x=df['Survived'],hue=df['Survived'])
```

```python
sns.countplot(x=df['Pclass'],hue=df['Pclass'])
```

b. Pie chart

```python
df['Survived'].value_counts().plot(kind='pie',autopct = '%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct = '%.2f')
```

Numerical Data

a.Histogram

```python
sns.histplot(df['Age'],bins=10)
```

b. Distplot

```python
sns.distplot(x=df['Age'],bins=20) #or sns.histplot(df['Age],kde=True)
```

c. Boxplot

```python
sns.boxplot(df['Age'])
```

```python
sns.boxplot(df['Fare'],orient='h',width=0.5)
```
