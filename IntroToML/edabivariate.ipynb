Bivariate Analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

1) Scatterplot -> Numerical - Numerical

```python
sns.scatterplot(x=iris['sepal_length'],y=iris['sepal_width'],hue=iris['species'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

2) Bar Plot -> Numerical - Categorical

```python
sns.barplot(x=titanic['pclass'],y=titanic['age'],hue=titanic['sex'])
```

3) Box Plot -> Numerical - Categorical

```python
sns.boxplot(x=titanic['age'],y=titanic['sex'],gap=0.5,hue=titanic['survived'])
```

4) Distplot -> Numerical - Categorical

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
```

5) Heatmap -> Categorical - Categorical

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']),annot=True)
```

```python
titanic.sample(1)
```

```python
titanic.groupby('pclass')['survived'].mean()
```

6) Clustermap -> Categorical - Categorical

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```

7) Pairplot

```python
sns.pairplot(iris,hue='species')
```

8) Lineplot -> Numerical - Numerical`

```python
flights.sample(5)
```

```python
sns.lineplot(x=flights['year'],y=flights['passengers'])
```

```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

```python
sns.heatmap(data=flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(data=flights.pivot_table(values='passengers',index='month',columns='year'))
```
