```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.axis([1,4,1,16])
plt.show()
```

```python
df=pd.read_csv("Iris.csv")
```

```python
df.head
```

```python
plt.bar(df["Species"],df['SepalLengthCm'])
plt.title("Bar plot")
plt.legend(['SepalLengthCm'])
plt.show()
```

```python
#histograms
plt.hist(df['SepalLengthCm'])
plt.title('Histogram')
plt.legend(['SepalLengthCm'])
plt.show()
```

```python
#scatter plot
plt.scatter(df['Species'],df['SepalLengthCm'])
plt.title("Scatter plot")
plt.legend(['SepalLengthCm'])
plt.show()
```

```python
#box plot
plt.boxplot(df['SepalWidthCm'])
plt.title('Box plot')
plt.legend(['SepalWidthCm'])
plt.show()
```

```python
#correlation heatmaps
num_df = df.drop(columns=['Species'])
plt.imshow(num_df.corr() , cmap = 'autumn' , interpolation = 'nearest' )
  
plt.title("Heat Map")
plt.show()
```
