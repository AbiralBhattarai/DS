```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Linear graph", fontsize=25, color="green")

# Adding label on the y-axis
plt.ylabel('Y-Axis')

# Adding label on the x-axis
plt.xlabel('X-Axis')

# Setting the limit of y-axis
plt.ylim(0, 80) 
plt.xlim(0,100)

# setting the labels of x-axis
plt.xticks(x,['one','two','three','four'])

plt.show()
```

3D plots in Matplotlib


```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0,1,100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)
ax.plot3D(x,y,z)
ax.set_title('3D plot')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0,1,100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)
c=x+y
ax.scatter(x,y,z,c=c)
ax.set_title('3D plot')
plt.show()
```

Data Visualization using Seaborn

```python
#Line plot
data = {'Name':['Mohe','Karnal','Abiral','Jack','Eddie'],
        'Age':[30,21,22,18,42],
        'Weight':[78,88,110,62,150]}
df = pd.DataFrame(data)
sns.lineplot(x=df['Name'],y=df['Weight'])
```

```python
#2 Scatter plots
sns.scatterplot(x=df['Age'],y=df['Weight'])
```

```python
#3 Box plots
data = {'Name':[ 'Mohe' , 'Karnal' , 'Yrik' , 'jack' ],
        'Age':[ 30 , 21 , 29 , 28 ]}
df = pd.DataFrame( data )
sns.boxplot(data['Age'])
```
