```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4])
y = x*x
plt.plot(x,y)
plt.title('X**2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
plt.figure(figsize=(10, 4))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time', style='time')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

Data visualization with matplotlib

```python
x = [1,2,3,4]
y=[1,5,8,12]
plt.plot(x,y)
plt.title('Simple line plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(
)
plt.show()
```

```python
data = sns.load_dataset('tips')
x = data['day']
y = data['total_bill']
plt.bar(x,y)
plt.xlabel('day')
plt.ylabel('Bill')
plt.title('Day vs Bill')
plt.show()
```

```python
x = data['total_bill']
plt.hist(x)
plt.title("Tips Dataset")

plt.ylabel('Frequency')

plt.xlabel('Total Bill')

plt.show()
```

```python
x = data['day']
y = data['total_bill']
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('Day')
plt.ylabel('Bill')
plt.show()
```

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]
plt.pie(data,labels=cars)
plt.title("Car data")

plt.show()
```

```python
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=False)

plt.xlabel('Data Set')
plt.ylabel('Values')
plt.title('Example of Box Plot')
plt.show()
```

```python
data = np.random.rand(10,10)
plt.imshow(data,interpolation='nearest')
```

```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Linear graph", fontsize=25, color="green")

# Adding label on the y-axis
plt.ylabel('Y-Axis')

# Adding label on the x-axis
plt.xlabel('X-Axis')

# Setting the limit of y-axis
plt.ylim(0, 80) 
plt.xlim(0,100)

# setting the labels of x-axis
plt.xticks(x,['one','two','three','four'])

plt.show()
```
